# GPU Process Size Trend

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

# CPU and GPU Die Size  Trend

```python
df_cpu_ps = df[['Type', 'Date', 'Process_Size']].where(df['Type'] == 'CPU').dropna()
df_cpu_ps['Date'] = pd.to_datetime(df_cpu_ps['Date'])
df_cpu_ps['Year'] = df_cpu_ps['Date'].dt.year
df_cpu_ps = df_cpu_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_cpu_ps.rename(columns={'Process_Size': 'CPU_Process_Size'}, inplace=True)

plt.plot(df_ps['Year'], df_ps['Process_Size'], color='green', label='GPU Process Size')
plt.plot(df_cpu_ps['Year'], df_cpu_ps['CPU_Process_Size'], color='blue', label='CPU Process Size')

plt.title('GPU and CPU Process Size Trend')
plt.ylabel('nm')
plt.xlabel('Year')
plt.legend()
plt.show()
```

# Conclusion:


Over the years, CPU and GPU process sizes have steadily decreased. Smaller process nodes generally result in faster, more efficient processors by shortening electrical signal travel distances and reducing power consumption.

```python
df_merged = df_ps.merge(df_cpu_ps, on='Year', how='inner')


df_merged.rename(columns={'Process_Size': 'GPU_Die_Size', 'CPU_Process_Size': 'CPU_Die_Size'}, inplace=True)


plt.plot(df_merged['Year'], df_merged['GPU_Die_Size'], color='red', label='GPU Die Size')
plt.plot(df_merged['Year'], df_merged['CPU_Die_Size'], color='orange', label='CPU Die Size')

plt.title('GPU and CPU Die Size Trend')
plt.ylabel('Die Size (nm)')
plt.xlabel('Year')
plt.legend()
plt.show()
```

# Conclusion

CPU and GPU die sizes have trended downward, reducing manufacturing costs and improving yield rates. This shrinkage also reflects advancements in transistor density, leading to enhanced performance.

# CPU and GPU Count Size Trend

```python

df_t = df_w[['Type', 'Date', 'Transistors']].where(df_w['Type'] == 'GPU').dropna()
df_t['Date'] = pd.to_datetime(df_t['Date'])
df_t['Year'] = df_t['Date'].dt.year
df_t = df_t.groupby('Year')['Transistors'].mean().reset_index()


df_cpu_t = df_w[['Type', 'Date', 'Transistors']].where(df_w['Type'] == 'CPU').dropna()
df_cpu_t['Date'] = pd.to_datetime(df_cpu_t['Date'])
df_cpu_t['Year'] = df_cpu_t['Date'].dt.year
df_cpu_t = df_cpu_t.groupby('Year')['Transistors'].mean().reset_index()
```

```python
df_combined = pd.merge(df_cpu_t, df_t, on='Year', how='inner', suffixes=('_CPU', '_GPU'))

plt.figure(facecolor='#F0F0D7', figsize=(10, 5))
plt.plot(df_combined['Year'], df_combined['Transistors_CPU'], color='#223C63', marker='', linestyle='-', label='CPU Transistor Count')
plt.plot(df_combined['Year'], df_combined['Transistors_GPU'], color='#F6724B', marker='', linestyle='-', label='GPU Transistor Count')
plt.title('CPU and GPU Transistor Count Trend')
plt.xlabel('Year')
plt.ylabel('Transistor Count')
plt.legend()
plt.grid(True)
plt.show()
```

# Average Frequency per Brand

```python
plt.figure(figsize=(10,5))
df.groupby('Vendor')['Frequency'].mean().plot(kind='bar', color=['#F6724B','#FC8A55','#8F4E51','#2B3349','#223C63'])
plt.title("Average Frequency per Brand")
plt.ylabel("GHz")
plt.xticks(rotation=45)
plt.show()
```

# Conclusion
The analysis of the average frequency per brand reveals notable differences among vendors. Intel stands out with the highest average frequency, significantly surpassing other brands. AMD also demonstrates a strong average, though notably lower than Intel. NVIDIA and ATI exhibit more moderate averages, while the "Other" category displays the lowest average frequency. This distribution suggests a clear performance differentiation, with Intel dominating in terms of frequency capabilities.
