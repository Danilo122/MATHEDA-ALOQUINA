import **pandas, numpy and matplotlib** library**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the file

```python
file_path = "Economic-Fully-Cleaned.csv"  
df = pd.read_csv(file_path)
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
plt.figure(figsize=(10, 5))
usa_data = df[df['Country'] == 'USA']
plt.plot(usa_data['Year'], usa_data['GDP (in billion USD)'], marker='o', linestyle='-', label='USA')
plt.xlabel('Year')
plt.ylabel('GDP (in billion USD)')
plt.title('GDP Over Time for USA')
plt.legend()
plt.grid(True)
plt.show()
```

# Conclusion

The GDP of the USA shows an overall increasing trend, indicating economic growth over the years. Some fluctuations may be observed, which could be attributed to economic recessions or periods of rapid growth.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='GDP (in billion USD)', y='Unemployment Rate (%)', hue='Country', alpha=0.7)
plt.xlabel('GDP (in billion USD)')
plt.ylabel('Unemployment Rate (%)')
plt.title('GDP vs Unemployment Rate')
plt.legend(loc='best', bbox_to_anchor=(1, 1))
plt.show()
```

# Conclusion

There appears to be a negative correlation between GDP and the unemployment rate; as GDP increases, the unemployment rate tends to decrease. 
Countries with higher GDP typically have lower unemployment rates, showing a stable labor market. 

```python
avg_gdp = df.groupby('Country')['GDP (in billion USD)'].mean().nlargest(10)
plt.figure(figsize=(10, 5))
avg_gdp.plot(kind='bar', color='skyblue')
plt.xlabel('Country')
plt.ylabel('Average GDP (in billion USD)')
plt.title('Top 10 Countries by Average GDP')
plt.xticks(rotation=45)
plt.show()
```

# Conclusion

The top 10 countries by average GDP likely include major economies such as the USA, China, Japan, and Germany. This chart helps highlight economic powerhouses and the disparities between different nations.

```python
plt.figure(figsize=(6, 4))
corr = df[['GDP (in billion USD)', 'Unemployment Rate (%)', 'Economic Growth (%)']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Economic Indicators')
plt.show()
```

# Conclusion

The heatmap visually confirms the relationships between GDP, Unemployment Rate, and Economic Growth. A strong positive correlation between GDP and economic growth suggests that as an economy grows, GDP increases. A negative correlation between unemployment and GDP aligns with economic theory when GDP rises, unemployment tends to fall.
